```python
from numpy import median #import required librairies
import matplotlib.pyplot as plt
```

```python
# This function plots the comparison between the runtimes of the lexicographical, graded lexicographical and the reverside graded lexicographical order. It takes as inputs the title, x_label, y_label, x_values, the runtimes and the maximal runtimes
def plotComparison(title, x_label, y_label, x_values, y1_values, y2_values, y3_values, y_max):
  plt.figure(figsize=(10, 6))  # Set plot size

  # Plot lines with labels
  plt.plot(x_values, y1_values, marker='o', label='lex', color='blue')
  plt.plot(x_values, y2_values, marker='o', label='deglex', color='green')
  plt.plot(x_values, y3_values, marker='o', label='degrevlex', color='red')

  # Set labels and title
  plt.xlabel(x_label)
  plt.ylabel(y_label)
  plt.title(title)

  # Set axis limits
  plt.xlim(min(x_values), max(x_values))
  plt.ylim(0,  y_max)
  plt.xticks(x_values)
  # Add legend
  plt.legend()

  # Show the plot
  plt.grid(False)
  plt.show()
```
