```python
from numpy import median #import required librairies
import matplotlib.pyplot as plt
```

```python
# This function plots the comparison between the runtimes of the solution of the extention and the solution in Q. It takes as inputs the title, x_label, y_label, x_values, the runtimes and the maximal runtimes

def plotComparison(title, x_label, y_label, x_values, y1_values, y2_values, y_max):
  plt.figure(figsize=(10, 6))  # Set plot size

  # Plot lines with labels
  plt.plot(x_values, y1_values, marker='o', label='rationals', color='black')
  plt.plot(x_values, y2_values, marker='x', label='cyclotomic', color='black')

  # Set labels and title
  plt.xlabel(x_label)
  plt.ylabel(y_label)
  plt.title(title)

  # Set axis limits
  plt.xlim(min(x_values), max(x_values))
  plt.ylim(0,  y_max)
  plt.xticks(x_values)
  # Add legend
  plt.legend()

  # Show the plot
  plt.grid(False)
  plt.show()
```

```python
#n = 7

p = [round(i, 2) for i in [x * 0.1 for x in range(1, 10)]] #Set ranging probabilities of edges between vertices

# The percieved runtimes.
OverQ = [0.02, 0.02, 0.07, 0.02, 0.03, 0.06, 0.06, 0.11, 0.11]
OverCyclotomic = [0.02, 0.01, 0.07, 0.02, 0.02, 0.03, 0.02, 0.6, 0.6]

title = r"$7$ vertices"
x_label = r"$\mathbb{P}($edge between vertices $i$ and $j$ s.t. $i \ne j)$"
y_label = r"seconds"

y_max = max(max(OverQ), max(OverCyclotomic))

plotComparison(title, x_label, y_label, p, OverQ, OverCyclotomic, y_max)

```
